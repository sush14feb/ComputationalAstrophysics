# Plotting in Python: Matplotlib

Visualisations have always been appealing as they provide ease in interpreting the data quickly. Whether its a function or some sort of distribution, a single look can reveal a lot of information regarding the data. Python offers easy plotting through its `matplotlib.pyplot` library and its associated functions.

```python
import matplotlib.pyplot as plt

#We also import numpy as it wil be of much use throughout
import numpy as np
```

The `plot()` function is used to generate the plots. You may check its arguments and outputs using `plt.plot?` command. Let us try plotting some basic data.

```python
plt.plot([1,2,3,4,5]) #The plot() function considers single argument as the 'y' data and plots it against 'x' which is an array from 0 to n-1
plt.show() #In Jupyter Notebooks the plot appears without the show() function but it becomes necessary if you are using any other editor
```

You may also give your own dataset for the $x$ axis, as shown below:

```python
plt.plot([0,1,2,3,4],[0,2,4,6,8]) #Here the first list is the 'x' data and the second list is the 'y' data
plt.show()
```

You can also do this by designating the lists/arrays as a variable and passing them through the `plt.plot()` function.

```python
x=np.array([0,1,2,3,4,5])
y=np.array([3,5,1,4,2,6])
plt.plot(x,y)
plt.show()
```

### Plot Appearance
By default, `plt.plot()` joins all the data points by a straight line. One can change this to a desirable line type. This is done by passing a third argument which is a string specifying the line type. The default is 'b-' which represents a blue line. The string is passed as the concatenation of color and type such as 'rs' would represent red ('r') squares('s'). 

```python
plt.plot(x,y,'rs')
plt.show()
```

```python
plt.plot(x,y,'g--') #Green broken line
plt.show()
```

An exhaustive lists of linetypes and color codes can be found <a href="https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html">here</a> 

You may also customize your markers using the `marker` argument of the `plt.plot()` function

```python
plt.plot(x,y,'g--', marker='o') #Green broken line with solid circles as markers
plt.show()
```

You are encouraged to explore its other properties such as size and face and edge colors designated by the keyword arguments `ms`, `mfc` and `mec`, respectively. <br>
Its important to describe your data by labelling the plot. The functions which come handy are the `plt.xlabel()` and `plt.ylabel()`. You can also give the plot a title.

```python
plt.plot(x,y,'g--')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Sample plot')
plt.show()
```

```python
#The text sizes can be adjusted as well
plt.plot(x,y,'g--', marker='o')
plt.xlabel('X Axis',size=10)
plt.ylabel('Y Axis',size=15)
plt.title('Sample plot',size=20)
plt.show()
```

In case you need a grid, you may use the `plt.grid()` function. <br> Now let us try plotting multiple dayasets in a single plot.

```python
x=np.array([0,1,2,3,4,5,6,7])
y1=x
y2=x**2
y3=3*x
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.show()
```

Let us make this plot more informative by customizing it and adding labels and a legend.

```python
x=np.array([0,1,2,3,4,5,6,7])
y1=x
y2=x**2
y3=3*x
plt.plot(x,y1,'b+',label='y=x')
plt.plot(x,y2,'gs',label='y=x^2')
plt.plot(x,y3,'ro',label='y=3x')
plt.legend() #Adds a legend at an appropriate place, you may fix its location using 'loc' argument
plt.xlabel('X Axis',size=15)
plt.ylabel('Y Axis',size=15)
plt.title('Sample Plot',size=20)
plt.grid()
plt.show()
```

You may also change the plot size using the following function:

```python
plt.figure(figsize=(8,6))
plt.plot(x,y1,'b+',label='y=x')
plt.plot(x,y2,'gs',label='y=x^2')
plt.plot(x,y3,'ro',label='y=3x')
plt.legend() #Adds a legend at an appropriate place, you may fix its location using 'loc' argument
plt.xlabel('X Axis',size=15)
plt.ylabel('Y Axis',size=15)
plt.title('Sample Plot',size=20)
plt.show()
```

## Going an extra step

`matplotlib` provides much more power to the user in terms of customizations. Let's look into the `subplot` utility which lets one creaye several subplots within a single figure.

```python
x=np.array([0, 1, 2, 3])
#plot 1:
y1=np.array([3, 2, 4, 1])
plt.subplot(1, 2, 1) #Can also be written as plt.subplot(121)
plt.plot(x,y1,'k-')

#plot 2:
y2=np.array([30, 20, 25, 10])
plt.subplot(1, 2, 2) #Can also be written as plt.subplot(122) 
plt.plot(x,y2,'b--')
plt.show()
```

If you write `plt.subplot(A,B,C)` where A, B and C are some numbers then it denotes that the figure as A rows and B columns and the given plot is the Cth plot. For example, the plots above have 1 row and 2 columns. The same can be done as shown below. Also you can add titles and axes labels for each plot.

```python
x=np.array([0, 1, 2, 3])

#plot 1:
y1=np.array([3, 2, 4, 1])
plt.subplot(2, 1, 1) #Can also be written as plt.subplot(211)
plt.plot(x,y1,'k-')

#plot 2:
y2=np.array([30, 20, 25, 10])
plt.subplot(2, 1, 2) #Can also be written as plt.subplot(212) 
plt.plot(x,y2,'b--')

plt.show()
```

Let's create 4 plots in a 2x2 grid in a single figure

```python
t=np.linspace(0,10,100)

plt.figure(figsize=(8,8))

plt.subplot(2, 2, 1) #1st row, 1st column
plt.plot(t, np.sin(t), 'g^') 

plt.subplot(2, 2, 2) #1st row, 2nd column
plt.plot(t, np.sin(t), 'b-')

plt.subplot(2, 2, 3) #2nd row, 1st column
plt.plot(t, np.sin(t), 'ko')

plt.subplot(2, 2, 4) #2nd row, 2nd column
plt.plot(t, np.sin(t), 'r--')

plt.show()
```

## Let's dwell into Astronomy: Plotting Constellations

For plotting the night sky, we need the coordinates of the stars. The coordinates of each star in a constellation has been scraped from <a href="https://en.wikipedia.org/wiki/Lists_of_stars_by_constellation">here</a>. The RA, Dec obtained are the coordinates on the celestial sphere which we project on the Cartesian plane using <a href="https://en.wikipedia.org/wiki/Stereographic_projection">stereographic projection</a>. The data for a few constellations has been scraped and can be found in the `csv` files in the same folder. The format of the file is `RA, Dec, Magnitude`. The <a href="https://en.wikipedia.org/wiki/Magnitude_(astronomy)">magnitude</a> of a star is measure of its brightness as observed from earth. We will use the apparent magnitude to create realistic constellations as one would normally see in the night sky.

```python
# Stereographic Projection. 
def stereo(ra, dec):
    theta = np.deg2rad(90-dec + dec.mean())
    phi = np.deg2rad((ra-ra.mean())*15)
    x = np.sin(theta)*np.cos(phi)
    y = np.sin(theta)*np.sin(phi)
    z = np.cos(theta)
    X = x/(1-z)
    Y = y/(1-z)
    return X,Y

#The plotting function
def plot_const(const_name, normalization=30):
    ra, dec, mag=np.loadtxt(f'{const_name}.csv',delimiter=',',unpack=True)
    x,y=stereo(ra,dec)
    s = 10**(-mag/2.5) #Controls the size of the marker proportional to the stellar magnitude
    s = s/s.max()*normalization #Normalized the sizes
    plt.figure(figsize=(10,10))
    plt.gca().set_facecolor('k')
    plt.scatter(-y,x,s=s,color='w') #Orientation can be changed as (x,y), (-y,x), (-x,-y) and (y,-x)
    plt.xticks([])
    plt.yticks([])
    plt.title(f'{const_name}',size=20)
```

```python
plot_const('Orion',normalization=50)
#Brownie points for identifying the stars, specially Pleiades
#NOTE: The stereopgraphic projection causes large distortion in the constellations near north pole
```
